```python
from matplotlib import animation, rc
rc('animation', html='html5')
```

```python
def h(theta0, theta1, x):
    return theta1*x + theta0

def mse(y_pred, y_true):
    err = (1/2) * (np.power((y_pred - y_true), 2)).mean()
    return err
```

```python
x, y = data[:, 0], data[:, 1]
```

```python
mu = x.mean()
```

```python
mu
```

```python
sigma = x.std()
sigma
```

```python
type(x)
```

$$x_{n} = \frac{x - \mu}{\sigma}$$

```python
xn = (x - mu) / sigma
```

```python
print(xn)
```

```python
#learning rate:
alpha = 5e-3

#initialize parameters randomly:
theta_0 = np.random.randn()
theta_1 = np.random.randn()

print("Initial guess:")
print(" theta_0 = {:.4f}\n theta_1 = {:5.4f}".format(theta_0, theta_1))

costs = []

def update_step():
    global theta_0, theta_1, costs
    
    #predict:
    y_pred = h(theta_0, theta_1, xn)
    costs.append(mse(y_pred, y))
    
    #compute gradiets:
    dtheta_0 = y_pred - y
    dtheta_1 = (y_pred - y) * xn
    
    #update parameters:
    theta_0 -= alpha * dtheta_0.mean()
    theta_1 -= alpha * dtheta_1.mean()
```

Animate

```python
fig = plt.figure(figsize=(5,3), dpi=140)

#plot data
plt.scatter(xn, y, s=30, marker='P')

#plot prediction (a line):
y_pred = h(theta_0, theta_1, xn)
line, = plt.plot(xn, y_pred, 'b')
plt.close

def animate(i):
    line.set_ydata(h(theta_0, theta_1, xn))
    for i in range(100): update_step()
    return line,
anim = animation.FuncAnimation(fig, animate, np.arange(0,20), interval=200, repeat_delay=1000)
```

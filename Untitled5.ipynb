```python
x = np.arange(0.001, 1, 0.001)
```

```python
plt.figure(figsize=(12, 4), dpi=200, edgecolor='green')

plt.subplot(131)
plt.plot(x, np.log(x))
plt.ylim(-5, 5)
plt.title(r'$\log(x)$')

plt.subplot(132)
plt.plot(x, -np.log(x))
plt.ylim(-5, 5)
plt.title(r'$-\log(x)$')

plt.subplot(133)
plt.plot(x, -np.log(1-x))
# plt.ylim(-5, 5)
plt.title(r'$-\log(1-x)$')

plt.show()
```

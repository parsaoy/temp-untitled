$$J(\theta) = \frac{1}{2} \sum_{i=1}^{m}(h_{\theta}(x^{(i)}) - y^{(i)})^{2}$$

```python
plt.style.use('seaborn-darkgrid')
```

```python
def plot_hyp(X, y, theta):
    plt.scatter(X[:,1], y, s=100, c='r', marker='*')
    
    x = np.linspace(0.8, 3.2)
    x = x[:, None]           # Increase Dimension by One
    x = np.hstack((np.ones_like(x), x))
    
    
    #plot hypothesis
    h = x @ theta
    plt.plot(x[:, 1], h, c='g', lw=3)
    plt.show()
```

```python
plot_hyp(X, y, theta)
```

```python
X = np.array([[1, 1],
              [1, 2],
              [1, 3]])

y = np.array([[1],
              [2],
              [3]])

theta = np.array([[1],
                  [2]])
```

```python
def sse(X, y, theta):
    
    predictions = X @ theta
    errors = predictions - y
    return 1/2 * (errors.T @ errors)[0,0]
```

```python
theta = np.array([[0],
                  [0]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J({}) = {:.2f}".format(theta.flatten(), loss))
```

```python
theta = np.array([[0], [1.5]])

plot_hyp(X, y, theta)
loss = sse(X, y, theta)
print("J({}) = {:.2f}".format(theta.flatten(), loss))
```

```python
theta_1s = np.linspace(-1, 3)
J = []
print(theta_1s)
```

```python
for theta_1 in theta_1s:
    cost = sse(X, y, np.array([[0],[theta_1]]))
    J += [cost]
```

```python
plt.figure(figsize=(12, 6), dpi=300)
plt.plot(theta_1s, J, color='green', linewidth=5)
plt.xlabel(r'$\theta_1$')
plt.ylabel(r"$J(\theta)$")

plt.annotate('Global Optimum', xy=(1,1), xycoords='data')
```

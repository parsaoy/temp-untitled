```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-darkgrid')
```

```python
x, step = np.linspace(-10, 10, 100, retstep=True)
```

```python
y = 1 / (1 + np.exp(-x))
```

```python
plt.figure(dpi=350, figsize=(12,5))
plt.plot(x, y, 'r^')
```

```python
x = np.arange(0.001, 1, 0.001)
```

```python
plt.figure(figsize=(12, 4), dpi=200, edgecolor='green')

plt.subplot(131)
plt.plot(x, np.log(x))
plt.ylim(-5, 5)
plt.title(r'$\log(x)$')

plt.subplot(132)
plt.plot(x, -np.log(x))
plt.ylim(-5, 5)
plt.title(r'$-\log(x)$')

plt.subplot(133)
plt.plot(x, -np.log(1-x))
# plt.ylim(-5, 5)
plt.title(r'$-\log(1-x)$')

plt.show()
```
